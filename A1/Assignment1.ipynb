```python

# Define the device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load the base pipeline
pipe = DiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",
    torch_dtype=torch.float16
).to(device)  # Move to GPU if available

# Load the LoRA weights
pipe.load_lora_weights("Fihade/Retro-Collage-Art-SDXL")

# Define the prompt
```

```python
prompt = "ff-collage, surreal people snowboarding down a hill in summer, with flowers, birds flying and a nice sun"
num_inference_steps = 20  # Adjust this value for quality vs. speed tradeoff

# Generate the image
image = pipe(prompt, num_inference_steps=num_inference_steps).images[0]

# Save or display the image
image
image.show()  # Display the image
image.save("retro_collage5.png")  # Save to file
```
